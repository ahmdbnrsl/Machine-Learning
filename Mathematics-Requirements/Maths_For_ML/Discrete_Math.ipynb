jika ada 10 orang dan saling bersalama-salaman (setiap orang hanya menyalami satu orang yang sama), maka berapa kali terjadi salam-salamannya?

```python
man = [chr(i) for i in range(97, 107)]

result = set()

def add(e, s):
    if e[::-1] not in s: s.add(e)

for i in man:
    for j in man:
        if j != i: add(i+j, result)

len(result)
```
