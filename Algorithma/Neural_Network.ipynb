```python
import numpy as np
np.random.seed(42)

x_train = np.array([[0], [8], [15]], dtype=float)
y_train = np.array([[32], [46.4], [59]], dtype=float)

w = np.random.randn(1, 1)
b = np.random.randn(1)
lr = 0.001

N = x_train.shape[0]

for i in range(40000):
    y_pred = np.dot(x_train, w) + b
    error_diff = y_train - y_pred

    mse = np.mean(error_diff ** 2)
    dw = -2 * np.dot(x_train.T, error_diff) / N

    db = -2 * np.mean(error_diff)

    w = w - (lr * dw)
    b = b - (lr * db)

w, b
```

```python
import numpy as np
np.random.seed(42)

x_train = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y_train = np.array([[0], [0], [0], [1]])

w = np.random.randn(2, 1)
b = np.random.randn()
learning_rate = 0.001

def sigmoid(z):
    return 1/(1 + np.exp(-z))

for i in range(500000):
    z = x_train @ w + b
    a = sigmoid(z)

    dw = (x_train.T @ (a - y_train)) / 4
    db = np.mean(a - y_train)

    w = w - (learning_rate * dw)
    b = b - (learning_rate * db)

def predict(x):
    return sigmoid(x @ w + b)

predict(np.array([[1, 1]]))
print(w, b)
```
