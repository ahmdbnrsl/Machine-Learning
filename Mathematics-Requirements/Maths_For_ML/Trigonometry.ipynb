**Contoh Soal Trigonometri**

Sebuah tangga dengan panjang $10$ meter bersandar pada dinding. Sudut antara tangga dan tanah adalah $60$. Berapakah ketinggian dinding yang dicapai oleh tangga?

langkah-langkah:

sisi miring segitiga = panjang tangga atau $10$ meter.

sudut bawah kanan = $60$ => sudut atas = $30$

gunakan fungsi $sin$

$sin(30) = \sqrt3 / 2$

$\sqrt3/2 = a/10$ (sisi miring)

$a = \sqrt3/2  10 = 5\sqrt3$

**Mengukur Tinggi Gunung dengan Trigonometri**

Seperti yang kita tahu bahwa $$\tan(\theta) = \frac{\text{depan}}{\text{samping}}$$

maka $$\tan(\theta) = \frac{h}{d}$$

nahh di ambil dari gambar diatas dapat disusun rumus sebagai berikut

$$\tan(\alpha) = \frac{h}{d}\\
\tan(\beta) = \frac{h}{d + x}$$
selesaikan dengan aljabar untuk mendapatkan $d$ dan $h$

$$d = \frac{h}{\tan(\alpha)}\\
d + x= \frac{h}{\tan(\beta)}\\
d = \frac{h}{\tan(\beta)} - x\\
\frac{h}{\tan(\beta)} - x = \frac{h}{\tan(\alpha)}\\
h = \frac{x\tan(\alpha)\tan(\beta)}{\tan(\alpha) - \tan(\beta)}$$
