**Rumus mengubah sudut dalam satuan derajat (degree) ke radians dan sebaliknya**

=> mengubah degree ke radians $\theta × (\pi/180°)$

contoh :

ubahlah $90°$ ke radians

$90° × (π/180°) = \pi / 2$

=> mengubah radians ke degree $\theta × (180/\pi)$

contoh :

ubahlah $2\pi$ ke degree

$2\pi × (180°/π) = 360°$
