```python

# AND LOGIC
import numpy as np
np.random.seed(42)

x_train = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y_train = np.array([[0], [0], [0], [1]])

w = np.random.randn(2, 1)
b = np.random.randn()
learning_rate = 0.001

def sigmoid(z):
    return 1/(1 + np.exp(-z))

for i in range(500000):
    z = x_train @ w + b
    a = sigmoid(z)

    dw = (x_train.T @ (a - y_train)) / 4
    db = np.mean(a - y_train)

    w = w - (learning_rate * dw)
    b = b - (learning_rate * db)

def predict(x):
    return sigmoid(x @ w + b)

predict(np.array([[1, 1]]))
print(w, b)
```


$$\boxed{\text{GRADIENT BACKWARD PROPAGATION XOR LOGIC}}$$

$$dw_1 = \frac{dE}{da_2}\cdot\frac{da_2}{dz_2}\cdot\frac{dz_2}{da_1}\cdot\frac{da_1}{dz_1}\cdot \frac{dz_1}{dw_1}$$

$$db_1 = \frac{dE}{da_2}\cdot\frac{da_2}{dz_2}\cdot\frac{dz_2}{da_1}\cdot\frac{da_1}{dz_1}\cdot \frac{dz_1}{dn_1}$$

$$dw_2 = \frac{dE}{da_2}\cdot\frac{da_2}{dz_2}\cdot\frac{dz_2}{dw_2}$$

$$db_2 = \frac{dE}{da_2}\cdot\frac{da_2}{dz_2}\cdot\frac{dz_2}{dw_2}$$

```python
# XOR LOGIC
import numpy as np
np.random.seed(42)

x_train = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y_train = np.array([[0], [1], [1], [0]])

def sigmoid(z):
    return 1/(1 + np.exp(-z))

w1 = np.random.randn(2, 2)
b1 = np.random.randn()

w2 = np.random.randn(2, 1)
b2 = np.random.randn()

learning_rate = 0.01

for i in range(50000):
    a1 = sigmoid(x_train @ w1 + b1)
    a2 = sigmoid(a1 @ w2 + b2)

    dw2 = (a1.T @ (a2 - y_train)) / 4
    db2 = np.mean(a2 - y_train)

    dz1 = ((a2 - y_train) @ w2.T) * (a1 * (1 - a1))
    dw1 = (x_train.T @ dz1)/ 4
    db1 = np.mean(dz1, axis=0)

    w1 = w1 - (learning_rate * dw1)
    w2 = w2 - (learning_rate * dw2)

    b1 = b1 - (learning_rate * db1)
    b2 = b2 - (learning_rate * db2)

def predict(x):
    a1 = sigmoid(x @ w1 + b1)
    a2 = sigmoid(a1 @ w2 + b2)

    return a2

predict(np.array([[1, 1]]))
```
