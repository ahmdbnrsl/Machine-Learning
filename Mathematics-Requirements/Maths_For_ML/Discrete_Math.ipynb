jika ada 10 orang dan saling bersalam-salaman (setiap orang hanya menyalami satu orang yang sama), maka berapa kali terjadi salam-salamannya?

```python
man = [chr(i) for i in range(97, 107)]

result = set()

def add(e, s):
    if e[::-1] not in s: s.add(e)

for i in man:
    for j in man:
        if j != i: add(i+j, result)

len(result)
```

**BASIC DISCRETE STRUCTURES**

**1. Sets**

Sets atau himpunan adalah sebuah kumpulan elemen yang tidak terurut, dan unik (artinya setiap ada pengulangan elemen yang sama, maka tidak dihitung)

contoh sets:
- a $$\{1, 2, 3, 4, 5\}$$
- b $$\{3, 1, 2, 5, 4\}$$
- c $$\left\{\sqrt{4}, 1, 8 - 3, \frac{24}{6}, \log_327, \log_4256\right\}$$

ketiga sets diatas dianggap sama, karena, memiliki kesamaan elemen walaupun tidak berurutan, dan walaupun ada pengulangan, karena pengulangan pada sets tidak dihitung.

contoh lain sets:

$$\{ x \in \mathbb{N}:x < 5, x≥1\}$$

apa maksud himpunan diatas? jadi itu adalah sebuah himpunan yang tidak dituliskan secara langsung tetapi kita memberitahu bahwa di dalam himpunan berisi sesuai kriteria yang di tuliskan, $x \in \mathbb{N}$ artinya element dari himpunan tersebut adalah anggota dari himpunan bilangan natural $\{0, 1, 2, 3, ...\}$, $x < 5, x≥ 1$ artinya x harus kurang dari 5 dan lebih dari atau samadengan 1.

- **union**

union adalah penggabungan 2 buah set, disimbolkan $A \cup B$, artinya setiap elemen yang ada di $A$ atau $B$ atau yang ada di keduanya.

- **intersection**

intersection berarti elemen yang sama atau yang terkandung di kedua set, disimbolkan $A \cap B$, yang berarti elemen yang ada di $A$ dan ada juga di $B$.

- **complement**

complement berarti setiap element yang tidak ada di himpunan tertentu, disimbolkan $A'$ yang berarti elemen yang tidak terkandung dalam himpunan $A$. Contoh: $A = \{x \in \mathbb{Z}\}$ dan $B = \left\{\frac{2}{3}, \frac{5}{2}\right\}$ maka bisa dikatakan bahwa $B$ adalah complement dari $A$

```python
A = {1, 2, 3}
B = {3, 4, 5}

# Union
union = A.union(B)
print(union)
# Intersection
intersection = A.intersection(B)
print(intersection)
```

**1.1. Subsets**

Subset adalah sebuah himpunan yang seluruh anggotanya merupakan anggota dari himpunan lain, disimbolkan $A\subseteq B$ artinya himpunan $A$ adalah subset atau samadengan himpunan $B$ contoh: $\{1, 2, 3\} \subseteq \{2, 1, 3\}$, $A \subsetneq B$ artinya himpunan $A$ subset dari $B$ tapi tidak sama dengan $B$ contoh: $\{1, 4\} \subsetneq \{1, 2 ,3, 4\}$, $A\not\subset B$ artinya $A$ bukanlah subset dari $B$ contoh: $\{0, 3\} \not\subset \{1, 2\}$.

```python
A = {1, 2, 3}
B = {1, 2, 3, 4, 5}

def subset(A, B):
    a, b = len(A), len(B)
    if a > b: return "Not Subset"
    elif a <= b:
        for i in A:
            if i not in B: return "Not Subset"
        return "A Subset B"

subset(A, B)
```
