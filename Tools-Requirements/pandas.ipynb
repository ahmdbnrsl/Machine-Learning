menghapus missing value pada dataframe (Nan)

```python
def make_missing_dataframe(nrows=30, ncols=4):
    data = np.random.randn(nrows, ncols)
    df = pd.DataFrame(data, columns=[cols[i] for i in range(ncols)])

    # Randomly assign NaN values
    for col in df.columns:
        df.loc[df.sample(frac=0.1).index, col] = np.nan

    return df

df = make_missing_dataframe()

# persentase missing value
df.isna().mean()
```

```python
df_backup = df.copy(deep=True)
print(df)

df = df.dropna(axis='columns')
df
```

```python
df = df_backup.copy(deep=True)
print(df)
df = df.dropna(axis='rows')
df
```

```python
print(df)
df = df_backup.copy(deep=True)
print(df)
treshold = len(df) * 0.01
df = df.dropna(thresh=treshold, axis='columns')
df
```

menggabungkan dua pandas data frame (merge)

```python
import pandas as pd
import numpy as np

n_rows = 10
n_cols = 5
cols = ('A', 'B', 'C', 'D', 'E')
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1, 4])
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0, 3])
df2
```

```python
# mengandung baris baris yang terkandung di kedua dataframe
df_inner = pd.merge(df1, df2, how='inner')
df_inner
```

```python
# menggabungkan 2 buah dataframe.
df_outer = pd.merge(df1, df2, how='outer')
df_outer
```

```python
import pandas as pd
import numpy as np

n_rows = 5
n_cols = 2

cols = ['omset', 'operasional']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df['omset'] = df['omset'] * 100_000
df['operasional'] = df['operasional'] * 10_000

df.index = pd.date_range(start='2025-05-08', periods=n_rows, freq='D')

df = df.reset_index()
df = df.rename(columns={'index': 'tanggal'})


formatku = {
    'tanggal': '{:%d-%m-%y}',
    'omset': 'Rp {:.2f}',
    'operasional': 'Rp {:.2f}'
}

laporan = df.style.format(formatku)

type(laporan)

laporan.set_caption('Laporan Omset dan Operasional')
laporan.highlight_min('omset', color='green')
laporan.highlight_max('omset', color='red')
laporan.highlight_min('operasional', color='green')
laporan.highlight_max('operasional', color='red')
```
